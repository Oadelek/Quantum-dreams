```python
# Import required modules
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import numpy as np

# Create a quantum circuit with one qubit
qc = QuantumCircuit(1)

# Apply a Hadamard gate to create a superposition state
#qc.h(0)
#qc.h(0)

qc.x(0)

# Define the rotation angle (e.g., pi/2 radians = 90 degrees)
theta = np.pi / 2

# Apply the RY gate to rotate around the Y-axis
qc.ry(theta, 0)

# Get the statevector from the circuit
state = Statevector.from_instruction(qc)

# Plot the Bloch sphere with a title and adjusted figure size
plot_bloch_multivector(state, title="Bloch Sphere for |+> State", figsize=(6, 6))
```

```python
from qiskit import QuantumCircuit
from qiskit.visualization import visualize_transition
import numpy as np

# RX Animation
qc_rx = QuantumCircuit(1)
qc_rx.h(0)
qc_rx.rx(np.pi/2, 0)
qc_rx.rz(np.pi/2, 0)
visualize_transition(qc_rx, trace=True, saveas="rx_rotation.mp4")
```

```python
# RY Animation
qc_ry = QuantumCircuit(1)
qc_ry.ry(np.pi/3, 0)
visualize_transition(qc_ry, trace=True, saveas="ry_rotation.mp4")
```

```python
# RZ Animation (with Hadamard first)
qc_rz = QuantumCircuit(1)
qc_rz.h(0)
qc_rz.ry(np.pi/2.25, 0)
qc_rz.rz(np.pi/4, 0)
visualize_transition(qc_rz, trace=True, saveas="rz_rotation.mp4")
```
